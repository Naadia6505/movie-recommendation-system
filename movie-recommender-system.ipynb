```python
actual_likes = {
    'User1': ['Inception', 'Tenet'],
    'User2': ['The Matrix', 'John Wick'],
    'User3': ['Interstellar', 'The Prestige']
}
```

```python
recommendations = {
    'User1': ['Interstellar', 'Tenet', 'Batman Begins', 'Dunkirk', 'Inception'],
    'User2': ['John Wick', 'The Matrix Reloaded', 'Speed', 'Constantine', '47 Ronin'],
    'User3': ['Tenet', 'Inception', 'Oppenheimer', 'The Dark Knight', 'The Prestige']
}
```

```python
def calculate_hit_rate(actual_likes, recommendations, k=5):
    hits = 0
    total_users = len(actual_likes)

    for user in actual_likes:
        actual = set(actual_likes[user])
        top_k_recs = recommendations[user][:k]
        hit = any(movie in actual for movie in top_k_recs)
        hits += int(hit)

    hit_rate = hits / total_users
    return round(hit_rate, 2)
```

```python
hit_rate_at_5 = calculate_hit_rate(actual_likes, recommendations, k=5)
print("🎯 Hit Rate @5:", hit_rate_at_5)
```

```python
def hit_rate(y_true, y_pred_top_n):
    hits = sum([1 if y in y_pred_top_n[i] else 0 for i, y in enumerate(y_true)])
    return (hits / len(y_true)) * 100

# Example:
y_true = ['Movie A', 'Movie B', 'Movie C']
y_pred_top_n = [
    ['Movie A', 'Movie D', 'Movie E'],
    ['Movie X', 'Movie B', 'Movie Y'],
    ['Movie F', 'Movie G', 'Movie H']
]

hr = hit_rate(y_true, y_pred_top_n)
print(f"Hit Rate: {hr:.2f}%")
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
import numpy as np

# Binary classification example
# 1 = Accept, 0 = Reject
y_true = [1, 1, 0, 1]  # Actual values
y_pred = [1, 1, 0, 0]  # Predicted values

# Confusion matrix components
TA = sum((np.array(y_true) == 1) & (np.array(y_pred) == 1))
FA = sum((np.array(y_true) == 0) & (np.array(y_pred) == 1))
FR = sum((np.array(y_true) == 1) & (np.array(y_pred) == 0))
TR = sum((np.array(y_true) == 0) & (np.array(y_pred) == 0))

# Calculate FAR and FRR
FAR = (FA / (FA + TR)) * 100 if (FA + TR) != 0 else 0
FRR = (FR / (FR + TA)) * 100 if (FR + TA) != 0 else 0

# Precision, Recall, F1-Score
precision = precision_score(y_true, y_pred) * 100
recall = recall_score(y_true, y_pred) * 100
f1 = f1_score(y_true, y_pred) * 100

# Print Results
print(f"True Accepts: {TA}")
print(f"False Accepts: {FA}")
print(f"False Rejects: {FR}")
print(f"True Rejects: {TR}")
print(f"False Acceptance Rate (FAR): {FAR:.2f}%")
print(f"False Rejection Rate (FRR): {FRR:.2f}%")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-Score: {f1:.2f}")
```
